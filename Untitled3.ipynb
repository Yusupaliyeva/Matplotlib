```python
import matplotlib.pyplot as plt
import numpy as np


x= np.array([1,2,6,8])
y = np.array([3,8,1,10])


plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


y = np.array([3,8,1,10,])


plt.plot(y,'o:r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


y = np.array([3,8,1,10,])


plt.plot(y,marker='o',ms=20)
plt.show()
```
